```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math 
import scipy.stats as stat
%matplotlib inline

from IPython.display import display

vtypes2012read = pd.read_csv('../DataSets/road_accident_data_by_vehicle_type_2012.csv')
vtypes2011read = pd.read_csv('../DataSets/road_accident_data_by_vehicle_type_2011.csv')
vtypes2010read = pd.read_csv('../DataSets/road_accident_data_by_vehicle_type_2010.csv')
vpopulation = pd.read_csv('../DataSets/vehicle_population.csv')
accidentsByPersonColombo = pd.read_csv('../DataSets/accidents_by_person_colombo.csv')
```

## Trim the unneccesary data from datasets 

```python
vtypes2012read = vtypes2012read.drop(columns=['Total'])
vtypes2011read = vtypes2011read.drop(columns=['Total'])
vtypes2010read = vtypes2010read.drop(columns=['Total'])
vtypes2012totals = vtypes2012read.loc[25:, 'Motor Car':'Other']
vtypes2011totals = vtypes2011read.loc[25:, 'Motor Car':'Other']
vtypes2010totals = vtypes2010read.loc[25:, 'Motor Car':'Other']

aggregetedTotals = vtypes2012totals.append(vtypes2011totals,ignore_index=True)
aggregatedTotals = aggregetedTotals.append(vtypes2010totals,ignore_index=True)
aggregatedTotalsTranspodsed = aggregatedTotals.T
aggregatedTotalsTranspodsed= aggregatedTotalsTranspodsed.reset_index()
aggregatedTotalsTranspodsed.columns = ['category','2012','2011','2010']
aggregatedTotalsTranspodsed
```

```python
#Setting the positions and width for the bars
pos = list(range(len(aggregatedTotalsTranspodsed['2012']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(30,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        aggregatedTotalsTranspodsed['2012'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=aggregatedTotalsTranspodsed['category'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        aggregatedTotalsTranspodsed['2011'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=aggregatedTotalsTranspodsed['category'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        aggregatedTotalsTranspodsed['2010'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=aggregatedTotalsTranspodsed['category'][2]) 

# Set the y axis label
ax.set_ylabel('accidents')

# Set the chart's title
ax.set_title('Accidents by vehicle type')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(aggregatedTotalsTranspodsed['category'],rotation=90)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(aggregatedTotalsTranspodsed['2010'] + aggregatedTotalsTranspodsed['2011'] + aggregatedTotalsTranspodsed['2012'])] )

# Adding the legend and showing the plot
plt.legend(['2010','2011','2012'], loc='upper left')
plt.grid()
plt.show()
```

```python
filteredPopulation2010 = vpopulation.loc[0:5,['vehicle class','2010']]
display(filteredPopulation2010)
```

```python
filteredPopulation2010.columns=['vehicle_type','population']
filteredPopulation2010['accidents'] = [11094,10054,16240,5989,10054,7924]
display(filteredPopulation2010)
```

```python
filteredPopulation2010['ratio'] = filteredPopulation2010.accidents/filteredPopulation2010.population
filteredPopulation2010.plot.bar(x='vehicle_type',y='ratio',figsize=(15,6))
plt.xlabel('vehicle category')
plt.ylabel('accident count as a ratio of vehicle population')
plt.title('Accidents/population ration by vehicle type')
```
