## Trim the unneccesary data from datasets 

```python
vtypes2012read = vtypes2012.drop(columns=['Total'])
vtypes2011read = vtypes2011.drop(columns=['Total'])
vtypes2010read = vtypes2010.drop(columns=['Total'])
vtypes2012totals = vtypes2012read.loc[25:, 'Motor Car':'Other']
vtypes2011totals = vtypes2011read.loc[25:, 'Motor Car':'Other']
vtypes2010totals = vtypes2010read.loc[25:, 'Motor Car':'Other']

aggregetedTotals = vtypes2012totals.append(vtypes2011totals,ignore_index=True)
aggregatedTotals = aggregetedTotals.append(vtypes2010totals,ignore_index=True)
aggregatedTotalsTranspodsed = aggregatedTotals.T
aggregatedTotalsTranspodsed= aggregatedTotalsTranspodsed.reset_index()
aggregatedTotalsTranspodsed.columns = ['category','2012','2011','2010']
aggregatedTotalsTranspodsed
```

```python
#Setting the positions and width for the bars
pos = list(range(len(aggregatedTotalsTranspodsed['2012']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(30,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        aggregatedTotalsTranspodsed['2012'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=aggregatedTotalsTranspodsed['category'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        aggregatedTotalsTranspodsed['2011'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=aggregatedTotalsTranspodsed['category'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        aggregatedTotalsTranspodsed['2010'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=aggregatedTotalsTranspodsed['category'][2]) 

# Set the y axis label
ax.set_ylabel('accidents')

# Set the chart's title
ax.set_title('Accidents by vehicle type')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(aggregatedTotalsTranspodsed['category'],rotation=90)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(aggregatedTotalsTranspodsed['2010'] + aggregatedTotalsTranspodsed['2011'] + aggregatedTotalsTranspodsed['2012'])] )

# Adding the legend and showing the plot
plt.legend(['2010','2011','2012'], loc='upper left')
plt.grid()
plt.show()
plt.savefig('../Visualizations/AccidentsByVehicleType.png')
```

#### improtant observations 

1. despite the low population numbers accidents ratio is high for districts `kilinochchi` and `matale`
2. `matara`,`kegalle` and `galle` has less population than `rathnapura` but the accident ratios are higher

```python
urbanPop = accidentsAndPopulation2012['urban_population']
median_urbanPop = urbanPop.median()
rural = accidentsAndPopulation2012[accidentsAndPopulation2012.urban_population < median_urbanPop]
rural_acc = rural['accidents']
urban = accidentsAndPopulation2012[accidentsAndPopulation2012.urban_population >  median_urbanPop]
urban_acc = urban['accidents']

rural_acc = rural_acc.sort_values()
urban_acc = urban_acc.sort_values()

print("realtively rural districts",rural['District'].count())
print("realtively urban districts",urban['District'].count())
```

```python
all_accidents = accidentsAndPopulation2012['accidents']
all_accidents = all_accidents.sort_values()
original_accidents = stat.norm.pdf(all_accidents, all_accidents.mean(), all_accidents.std())
plt.plot(all_accidents, original_accidents, label='All accidents')

urban_accidents = stat.norm.pdf(urban_acc, urban_acc.mean(), urban_acc.std())
plt.plot(urban_acc, urban_accidents, label='Urban accidents')

rural_accidents = stat.norm.pdf(rural_acc, rural_acc.mean(), rural_acc.std())
plt.plot(rural_acc, rural_accidents, label='Rural accidents')

plt.title('Distribution of accidents ')
plt.xlabel('accidents')
plt.ylabel('pdf')
plt.legend()
plt.savefig('../Visualizations/DistributionOfAccidentsPDF.png')
```

### Hypotheses 

1. Urban areas are more accident prone than rural areas
    - H0: both rural and urban areas have simillar chances for accidents
    - H1: urban areas have higher chance of accidents than rural areas
    
### Testing the hypotheses

- method: t test

```python
stats, p = stat.ttest_ind(urban_accidents, rural_accidents)
print('Statistics=%.3f, p=%.3f' % (stats, p))

alpha = 0.05  # 5% area under the normal graph
if p > alpha:
	print('Both urban and rural areas are similarly prone to accidents')
else:
	print('Urban areas are more prone to accidents than ruaral(reject H0)')
```
